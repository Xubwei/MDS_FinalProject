```python
rent_df['date'] = rent_df['rent_time'].dt.date
rent_df['hour'] = rent_df['rent_time'].dt.hour
group_rent = rent_df.groupby(['rent_station_id', 'date', 'hour']).count().reset_index()
group_rent = group_rent.iloc[:,:4]
group_rent.rename(columns={'rent_station_id':'station_id', 'rent_time':'count'}, inplace=True)
group_rent
```

```python
#把每小時資料補齊並填上0
from itertools import product
all_dates = pd.date_range(group_rent['date'].min(), group_rent['date'].max(), freq='D')
all_hours = range(24)
rent_stations = group_rent['station_id'].unique()
all_combinations = pd.DataFrame(product(rent_stations, all_dates, all_hours), columns=['station_id', 'date', 'hour'])

# 將 date 和 hour 合併成一個 datetime 欄位
all_combinations['date'] = pd.to_datetime(all_combinations['date'])
all_combinations['time'] = all_combinations.apply(lambda row: row['date'] + pd.Timedelta(hours=row['hour']), axis=1)
group_rent['date'] = pd.to_datetime(group_rent['date'])
merged_rent = pd.merge(all_combinations, group_rent, on=['station_id', 'date', 'hour'], how='left').fillna({'count': 0})

# 丟棄不再需要的 date 和 hour 欄位
merged_rent.drop(['date', 'hour'], axis=1, inplace=True)
merged_rent
```

```python
# 把每10分鐘補進來，並且把每小時的資料除以6
extend_rent = pd.DataFrame()

for index, row in merged_rent.iterrows():
    new_row = row.copy()
    new_row['count'] /= 6
    extend_rent = extend_rent.append(new_row, ignore_index=True)
    # 對於每一行，創建 5 個新行，對應於 10、20、30、40、50 分
    for minute in [10, 20, 30, 40, 50]:
        new_row = row.copy()
        new_row['count'] /= 6
        new_row['time'] = new_row['time'].replace(minute=minute, second=0)
        extend_rent = extend_rent.append(new_row, ignore_index=True)
extend_rent   
```

### 補資料

```python
import pandas as pd
import numpy as np
df = pd.read_csv('./chained_data(把time欄位給drop掉就可以用了).csv')
df['time'] = pd.to_datetime(df['time'])
```

```python
guan_1 = df[df['公館1號'] == 1]
guan_2 = df[df['公館2號'] == 1]
guan_3 = df[df['公館3號'] == 1]
guan_4 = df[df['公館4號'] == 1]
```

```python
#用forward fill跟backward fill的平均來填補缺失值
def mean_fill(f,b):
    mean_f = []
    for i in f.index:
        if math.isnan(f[i]):
            mean_f.append(b[i])
        elif math.isnan(b[i]):
            mean_f.append(f[i])
        else:
            mean_f.append((f[i]+b[i])/2)
    return mean_f
for col in ['進站人數', '出站人數', '降水量(mm)', '風向(360degree)']:
    f_fill = guan_1[col].fillna(method='ffill')
    b_fill = guan_1[col].fillna(method='bfill')
    m_fill = mean_fill(f_fill,b_fill)
    guan_1[col] = m_fill
    f_fill = guan_2[col].fillna(method='ffill')
    b_fill = guan_2[col].fillna(method='bfill')
    m_fill = mean_fill(f_fill,b_fill)
    guan_2[col] = m_fill
    f_fill = guan_3[col].fillna(method='ffill')
    b_fill = guan_3[col].fillna(method='bfill')
    m_fill = mean_fill(f_fill,b_fill)
    guan_3[col] = m_fill
    f_fill = guan_4[col].fillna(method='ffill')
    b_fill = guan_4[col].fillna(method='bfill')
    m_fill = mean_fill(f_fill,b_fill)
    guan_4[col] = m_fill
```

```python
guan_1['tot'] = guan_1['tot'].fillna(method='ffill')
guan_2['tot'] = guan_2['tot'].fillna(method='ffill')
guan_3['tot'] = guan_3['tot'].fillna(method='ffill')
guan_4['tot'] = guan_4['tot'].fillna(method='ffill')
```

```python
def mean_fill(df):
    sbi_list = []
    act_list = []
    for index, row in df.iterrows():
        if pd.isna(row['sbi']):
            missing_date = row['time']
            # 計算同一 weekday 和時間點的前三週數據的平均值
            count = 0
            cum_sbi = 0
            for i in range(1,10):
                if len(df[df['time'] == missing_date - pd.DateOffset(weeks=i)]) != 0 and ~np.isnan(df[df['time'] == missing_date - pd.DateOffset(weeks=i)]['sbi'].values[0]):
                    cum_sbi += df[df['time'] == missing_date - pd.DateOffset(weeks=i)]['sbi'].values[0]
                    count += 1
                if len(df[df['time'] == missing_date + pd.DateOffset(weeks=i)]) != 0 and ~np.isnan(df[df['time'] == missing_date + pd.DateOffset(weeks=i)]['sbi'].values[0]):
                    cum_sbi += df[df['time'] == missing_date + pd.DateOffset(weeks=i)]['sbi'].values[0]
                    count += 1
            if count == 0:
                average_value = 0
            else:
                average_value = round(cum_sbi/count)
            sbi_list.append(average_value)
        else:
            sbi_list.append(row['sbi'])
        if pd.isna(row['act']):
            missing_date = row['time']
            # 計算同一 weekday 和時間點的前三週數據的平均值
            one_count = 0
            zero_count = 0
            for i in range(1,10):
                if len(df[df['time'] == missing_date - pd.DateOffset(weeks=i)]) != 0:
                    if df[df['time'] == missing_date - pd.DateOffset(weeks=i)]['act'].values[0] == 1:
                        one_count += 1
                    else:
                        zero_count += 1
                if len(df[df['time'] == missing_date + pd.DateOffset(weeks=i)]) != 0:
                    if df[df['time'] == missing_date + pd.DateOffset(weeks=i)]['act'].values[0] == 1:
                        one_count += 1
                    else:
                        zero_count += 1
            if one_count > zero_count:
                average_value = 1
            else:
                average_value = 0
            act_list.append(average_value)
        else:
            act_list.append(row['act'])
    df['sbi'] = sbi_list
    df['act'] = act_list
    return df
guan_1 = mean_fill(guan_1)
guan_2 = mean_fill(guan_2)
guan_3 = mean_fill(guan_3)
guan_4 = mean_fill(guan_4)
df = pd.concat([guan_1, guan_2, guan_3, guan_4], axis=0).reset_index().drop('index', axis=1)
df['bemp'] = df['tot'] - df['sbi']
```

```python
df.to_csv("./chained_data(補完值).csv", index=False, encoding='utf-8-sig')
```
