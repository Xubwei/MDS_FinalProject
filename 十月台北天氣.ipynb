```python
import pandas as pd
from datetime import datetime, timedelta
import seaborn as sns
import matplotlib.pyplot as plt
```

##### concat all dfs together

```python
# 設定開始日期和結束日期
start_date = datetime(2023, 10, 1)
end_date = datetime(2023, 10, 31)

# 初始化一個空的 DataFrame 以便存放所有資料
october = pd.DataFrame()

# 迴圈讀取每一天的檔案
current_date = start_date
while current_date <= end_date:
    # 產生檔案名稱
    file_name = f"466920-{current_date.strftime('%Y-%m-%d')}.csv"
    
    # 讀取 CSV 檔案
    df = pd.read_csv(file_name)

    # 刪除每個日期的第一行文字
    df = df.iloc[1:]
    
    # 新增一個名為 "date" 的欄位，並將日期指派給每一個 row
    df['date'] = current_date.strftime('%Y-%m-%d')

    # 將當前讀取的資料串接到 all_data DataFrame 中
    october = pd.concat([october, df])
    
    # 移動到下一天
    current_date += timedelta(days=1)
    print(file_name)
# 顯示合併後的資料
print(october)
```

```python
# 設定子圖的列數和行數
n_rows = 6
n_cols = 3

# 創建子圖的配置
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 35))
axes_flat = axes.flatten()

# 繪製每個變數的直方圖
for i, column in enumerate(october.columns):
    if column != 'date':  # 不繪製日期欄位的直方圖
        sns.histplot(october[column], bins=30, ax=axes_flat[i])
        axes_flat[i].set_title(column, fontsize=10)
        axes_flat[i].set_xlabel('')
        axes_flat[i].set_ylabel('')

# 移除多餘的子圖
for ax in axes_flat[len(october.columns):]:
    fig.delaxes(ax)

plt.tight_layout()
plt.show()
```
